##pH measurements of M2 cells using a SNARF Biosensor

```python
'''M2 cells are human melanoma cells that have a deficiency in the expression of the filamin A gene. One of the phenotypes of M2 cells is their tendency to form blebs uncontrollably.
At the same time, cells use blebs for directed motility, implying that they can control the location of bleb formation.
While it is known that disruption of the actin cortex results in a similar phenotype, we hypothesized that bleb nucleation also has a physiological component 
originating from an efflux of protons and an influx of calcium ions at the cell periphery as in Yi et al. JBC. 2012. In this dataset, I will show you data from the pH sensor SNARF-1
'''
```
